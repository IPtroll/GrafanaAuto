```python
# for file in os.listdir("library"):

for file in os.listdir("library"):
    obj = json.loads(open(f"library/{file}").read())
    # check if exists
    print(obj["uid"])
    try:
        exist = LibraryElement(grafana.client).get_library_element(obj["uid"])
        print("exists")
        obj["version"] = exist["result"]["version"]
        print(LibraryElement(grafana.client).update_library_element(obj))
        print("updated")
        continue   
    except GrafanaClientError as e:
        print(e)
        pass

    try:
        exists = LibraryElement(grafana.client).get_library_element_by_name(obj["name"])["result"][0]
        print("exists", exists)
        print(LibraryElement(grafana.client).update_library_element({"uid": exists["uid"], "name": exists["name"]+"-old"}))
        print("renamed")
    except GrafanaClientError as e:
        pass
    
    LibraryElement(grafana.client).create_library_element(obj)
```
