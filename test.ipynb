```python


orgs = grafana.organizations.list_organization()
# print(json.dumps(orgs, indent=2))
for org in orgs:
    if org["name"] == "Main Org.": continue
    print(org)
    grafana.organizations.switch_organization(org["id"])
    for file in os.listdir("library/plots"):
        uid = file.replace(".json", "")
        res = create_library_element(grafana, uid)
        print(f"{org['name']}({org['id']}) {uid}: {res}")
```

```python
# update plots for org
org_name = "V1"

orgs = grafana.organizations.list_organization()
org = [org for org in orgs if org["name"] == org_name][0]
grafana.organizations.switch_organization(org["id"])

for file in os.listdir("library/plots"):
    uid = file.replace(".json", "")
    res = update_library_element(grafana, uid)
    print(f"{uid}: {res}")
```

```python
# create datasources for each org
orgs = grafana.organizations.list_organization()
for org in orgs:
    for file in os.listdir("library/data_sources"):
        uid = file.replace(".json", "")
        res = create_datasource(grafana, uid)
        print(f"{org['name']}({org['id']}) {uid}: {res}")
```
